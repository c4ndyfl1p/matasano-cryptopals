```python
def do_xor(pt_l, key_l):
    key_int = [ord(i) for i in key_l]
    ct = []
    pt_int = [ord(i) for i in pt_l]
    
    for i in range(len(pt_l)):
        ct.append  (hex(( pt_int[i] ^ key_int[i%3] ))[2:])
    return "".join(ct)

ct = do_xor(pt,key)
print(ct)

ans = "0b3637272a2b2e63622c2e69692a23693a2a3c6324202d623d63343c2a26226324272765272a282b2f20430a652e2c652a3124333a653e2b2027630c692b20283165286326302e27282f"

if ct == ans:
    print("success!")
else:
    print("wrong")
```
